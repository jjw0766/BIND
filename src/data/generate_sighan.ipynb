```python
import json
import pandas as pd
import datasets
```

```python
sighan_13_ds = datasets.load_dataset('AnonymousSubmissionOnly/sighan13')
sighan_14_ds = datasets.load_dataset('AnonymousSubmissionOnly/sighan14')
sighan_15_ds = datasets.load_dataset('AnonymousSubmissionOnly/sighan15')
```

```python
with open("C-LLM/train/wang_cscd_train.json", "r", encoding='utf-8') as f:
	wang_cscd_train = json.load(f)
```

```python
sentences_noisy = []
sentences = []
categories = []
```

```python
for row in wang_cscd_train:
    sentence_noisy, sentence, category = row['src'], row['tgt'], 'wang_cscd'
    if len(sentence)!=len(sentence_noisy):
        raise ValueError('sentence length not matched')
    sentences_noisy.append(sentence_noisy)
    sentences.append(sentence)
    categories.append(category)
```

```python
for d in sighan_13_ds['train']:
    category = 'sighan13'
    sentence_noisy, sentence = d['text'].split('\t')
    sentences_noisy.append(sentence_noisy)
    sentences.append(sentence)
    categories.append(category)

for d in sighan_14_ds['train']:
    category = 'sighan14'
    sentence_noisy, sentence = d['text'].split('\t')
    sentences_noisy.append(sentence_noisy)
    sentences.append(sentence)
    categories.append(category)

for d in sighan_15_ds['train']:
    category = 'sighan15'
    sentence_noisy, sentence = d['text'].split('\t')
    sentences_noisy.append(sentence_noisy)
    sentences.append(sentence)
    categories.append(category)
```

```python
train_df = pd.DataFrame({
    'sentence': sentences,
    'sentence_noisy': sentences_noisy,
    'category': categories
})  
train_ds = datasets.Dataset.from_pandas(train_df)
```

```python
sentences_noisy = []
sentences = []
categories = []

for d in sighan_13_ds['test']:
    category = 'sighan13'
    sentence_noisy, sentence = d['text'].split('\t')
    sentences_noisy.append(sentence_noisy)
    sentences.append(sentence)
    categories.append(category)

for d in sighan_14_ds['test']:
    category = 'sighan14'
    sentence_noisy, sentence = d['text'].split('\t')
    sentences_noisy.append(sentence_noisy)
    sentences.append(sentence)
    categories.append(category)

for d in sighan_15_ds['test']:
    category = 'sighan15'
    sentence_noisy, sentence = d['text'].split('\t')
    sentences_noisy.append(sentence_noisy)
    sentences.append(sentence)
    categories.append(category)
```

```python
test_df = pd.DataFrame({
    'sentence': sentences,
    'sentence_noisy': sentences_noisy,
    'category': categories
})  
test_ds = datasets.Dataset.from_pandas(test_df)
```

```python
ds = datasets.DatasetDict({
    'train': train_ds,
    'test': test_ds
})
```

```python
ds.push_to_hub(repo_id=f'jwengr/sighan', private=False)
```
